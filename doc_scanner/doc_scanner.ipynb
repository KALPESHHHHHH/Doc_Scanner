```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
# image path
img_path=r"C:\Projects\DS Projects\doc_scanner\img\doc_scanner.png"

# to read image 
img = cv2.imread(img_path)
print(img.shape)

#resizing the image
img=cv2.resize(img,(1500,1000))
print(img.shape)

#showing the image 
plt.imshow(img)
plt.show()
```

#Removing the noise
#Edge Detection 
#Contour Extraction
#Best Contour Selection
#Project to the screen

```python
# Removing noise - Blur the image 

org = img.copy()
gray=cv2.cvtColor(org,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
plt.show()

blurred= cv2.GaussianBlur(gray,(5,5),0)
plt.imshow(blurred,cmap="binary")
plt.show()
```

```python
gen_to_org=cv2.cvtColor(blurred,cv2.COLOR_GRAY2BGR)
plt.imshow(org)
plt.show()

plt.imshow(gen_to_org)
plt.show()
```

```python
gen_to_org.shape
```

```python
#Edge Detaction 
edge=cv2.Canny(blurred,0,50)
org_edge=edge.copy()
plt.title("Edge Detection")
plt.imshow(org_edge)
plt.show()
```

```python
#Countour Extraction

countours, _=cv2.findContours(edge,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
print(len(countours))

countours=sorted(countours,reverse=True,key=cv2.contourArea)
```
